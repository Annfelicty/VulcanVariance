```python
#Importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# reading the final_df 
df = pd.read_csv('final_df.csv')
df.head()
```

```python
# checking for the null values, from the highest to the lowest %
(df.isna().mean()*100).sort_values(ascending = False)
```

```python
# checking the skewness of the columns with missing data
# to determine how best to deal with the missing data

df['popularity'].skew()
```

```python
df['tmdb_votes'].skew()
```

```python
df['tmdb_rating'].skew()
```

```python
# we are going to drop the production_budget as it has the highest
# % of missing value (43%), plus the data is quite similar to the 
# production_budget_imputed 

df = df.drop(columns=['production_budget'])


# Fill missing values using the median for both popularity and tmdb_votes
# we assume the data is missing, and not an error
df['popularity'] = df['popularity'].fillna(0)
df['tmdb_votes'] = df['tmdb_votes'].fillna(0)

# filling with the mean becasue tmdb is nearly normally distributed
df['tmdb_rating'] = df['tmdb_rating'].fillna(df['tmdb_rating'].mean())
```

```python
# Distribution of the imdb-rating
sns.histplot(df['imdb_rating'], bins=30, kde=True)
plt.title("IMDb Ratings")

print(df['imdb_rating'].skew())
```

*Observarion:* The imdb_rating is strongly left skewed, suggesting there are outliers below the mean. 

```python
# Distribution of the tmdb_rating 

sns.histplot(df['tmdb_rating'], bins=30, kde=True)
plt.title("TMDb Rating")

print(df['tmdb_rating'].skew())
```

*Observation:*The tmdb_rating is nearly normally distributed. Majority of the data is near the mean.

```python
# Rating Distributions (IMDB vs TMDB)
plt.figure(figsize=(10, 4))
sns.kdeplot(df['imdb_rating'], label='IMDB Rating', fill=True)
sns.kdeplot(df['tmdb_rating'], label='TMDB Rating', fill=True)
plt.title('Comparison of IMDB and TMDB Rating Distributions')
plt.xlabel('Rating')
plt.ylabel('Density')
plt.legend()
plt.show()
```

*Observation:* Comparison of IMDB and TMDB Rating Distributions.

```python
# Genre counts
genre_counts = df["main_genre"].value_counts()
plt.figure(figsize=(6,4))
sns.barplot(y=genre_counts.index, x=genre_counts.values, orient="h")
plt.title("Top Genres")
plt.show()
```

*Observation:* the top genres are action and comedy

```python
# Yearly counts
plt.figure(figsize=(7,3))
sns.countplot(data=df, x="year_only", palette="viridis")
plt.title("Releases per Year")
plt.show()
```

*Observation:* 2016 had the highest movies released.

```python
# Top 10 studios
studio_counts = df["studio"].value_counts().head(10)
plt.figure(figsize=(6,4))
sns.barplot(y=studio_counts.index, x=studio_counts.values, orient="h")
plt.title("Top Studios by Release Count")
plt.xlabel("Number of Movies")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

*Observation:* the barplot shows the top 15 studio by the release count

```python
# runtime distribution
plt.figure(figsize=(6,4))
sns.histplot(df['runtime_minutes'], bins=30, kde=True)
plt.title('Distribution of Movie Runtimes')
plt.xlabel('Runtime (minutes)')
plt.ylabel('Frequency')
plt.show()

print(df['runtime_minutes'].skew())
```

*Observation:* 
The runtime is right-skewed. This means that some movies have longer than
the mean runtime. 
    

BIVARIABLE 

```python
#Bar plots
# Average gross by main genre
plt.figure(figsize=(10, 4))
df.groupby('main_genre')['total_gross'].mean().sort_values(ascending=False).plot(kind='bar')
plt.title('Average Total Gross by Main Genre')
plt.ylabel('Average Total Gross')
plt.xticks(rotation=45)
plt.show()
```

*Observation:* 

Adventure movie generate the highest revenue

```python
# Average rating by year
plt.figure(figsize=(10, 4))
df.groupby('year_only')['imdb_rating'].mean().plot(kind='line', marker='o')
df.groupby('year_only')['tmdb_rating'].mean().plot(kind='line', marker='o')
plt.title('Average IMDB Rating by Year')
plt.ylabel('Average Rating')
plt.grid()
plt.legend()
plt.show()
```

*Observation:* 

2014 & 2016 had the highest average imdb_rating

2018 had the highest rating for tmdb_rating

```python
# Relationships between financial metrics and ratings
pairs = [('imdb_rating', 'total_gross'),
         ('tmdb_rating', 'total_gross'),
         ('runtime_minutes', 'total_gross')]

for x, y in pairs:
    plt.figure(figsize=(6, 3))
    sns.scatterplot(data=df, x=x, y=y, alpha=0.6)
    plt.title(f'{y} vs {x}')
    plt.show()
```

*Observation:* 

Clustering 
Distribution

MULTIVARIATE

```python
# Proportion of genres(TOP 5) by year
ct = pd.crosstab(df['year_only'], df['main_genre'], normalize='index')
ct[ct.columns[:5]].plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Genre Proportion by Year (Top 5 Genres)')
plt.ylabel('Proportion')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1))
plt.show()
```

```python
#Time Series Relationships
# Plot financial metrics over time
plt.figure(figsize=(14, 7))
df.groupby('year_only')[['domestic_gross', 'foreign_gross', 'total_gross']].mean().plot()
plt.title('Average Gross Earnings Over Time')
plt.ylabel('Average Gross Earnings ($)')
plt.grid()
plt.show()
```

```python
# Plot budgets over time
plt.figure(figsize=(10, 5))
df.groupby('year_only')['production_budget_imputed'].mean().plot()
plt.title('Average Production Budget Over Time')
plt.ylabel('Average Budget ($)')
plt.grid()
plt.show()
```

```python
# RATING VS POPULARITY
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='imdb_rating', y='popularity', alpha=0.6)
plt.title('Popularity vs IMDB Rating')
plt.xlabel('IMDB Rating')
plt.ylabel('Popularity')
plt.show()
```

```python
#Heatmap of grouped average
# Calculate average metrics by genre and year
pivot_table = df.pivot_table(index='main_genre', 
                            columns='year_only', 
                            values='total_gross', 
                            aggfunc='mean')

plt.figure(figsize=(8, 4))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=False)
plt.title('Average Total Gross by Genre and Year')
plt.xlabel('Year')
plt.ylabel('Genre')
plt.show()
```
